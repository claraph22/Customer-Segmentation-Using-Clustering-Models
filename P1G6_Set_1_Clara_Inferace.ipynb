# Import Libraries

```python
import pandas as pd
import pickle
```

# Load Model

```python
with open('model_scaler.pkl', 'rb') as file_1:
    model_scaler = pickle.load(file_1)

with open('model_kmeans.pkl', 'rb') as file_2:
    model_kmeans = pickle.load(file_2)
with open('model_pca.pkl', 'rb') as file_3:
    model_pca = pickle.load(file_3)
```

```python
data_inf = pd.DataFrame([{
    'NAME': 'Clara',
    'BALANCE':8949 ,
    'BALANCE_FREQUENCY':1 ,
    'PURCHASES': 41050 ,
    'ONEOFF_PURCHASES':40624 ,
    'INSTALLMENTS_PURCHASES':14686,
    'CASH_ADVANCE':47137,
    'PURCHASES_FREQUENCY':100000,
    'ONEOFF_PURCHASES_FREQUENCY':1 ,
    'PURCHASES_INSTALLMENTS_FREQUENCY':1,
    'CASH_ADVANCE_FREQUENCY':1 ,
    'CASH_ADVANCE_TRX': 123,
    'PURCHASES_TRX':347 ,
    'CREDIT_LIMIT':30000 ,
    'PAYMENTS': 50721 ,
    'MINIMUM_PAYMENTS': 76407,
    'PRC_FULL_PAYMENT': 1,
    'TENURE': 12
}])
```

```python
data_inf
```

# SCALING

```python
data_inf.drop(['NAME'],axis=1,inplace=True)
```

```python
inf_scaled=model_scaler.transform(data_inf)
```

# PCA

```python
inf_pca=model_pca.transform(inf_scaled)
```

# Predict cluster

```python
inf_cluster=model_kmeans.predict(inf_pca)

if inf_cluster==0:
    inf_cluster='High-Balance, Low-Payment'
    print('High-Balance, Low-Payment')
elif inf_cluster==1:
    inf_cluster='Moderate-Balance, Moderate-Payment'
    print('Moderate-Balance, Moderate-Payment')
elif inf_cluster==2:
    inf_cluster='High-Balance, High-Payment'
    print('High-Balance, High-Payment')
```

didapatkan cluster `Moderate-Balance, Moderate-Payment` untuk data inferance ini
